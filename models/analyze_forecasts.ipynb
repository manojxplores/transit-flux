```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df_lgbm = pd.read_csv('../lgbm_model_metrics.csv')
```

```python
df_lgbm.describe()
```

```python
for metric in ['test_MAE', 'test_sMAPE']:
    fig = px.box(
        df_lgbm,
        x='model',
        y=metric,
        color='model', 
        color_discrete_map={
    'LightGBM': '#1f77b4',
    'XGBoost': '#ff7f0e' 
    },
    )
    
    fig.update_layout(
        title=f'Model Comparison: {metric} Distribution',
        xaxis_title='Model',
        yaxis_title=f'{metric}',
        font=dict(size=12)
    )
    
    fig.show()
```
